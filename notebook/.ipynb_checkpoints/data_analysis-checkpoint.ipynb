## 原始数据分析
数据说明：在初赛和复赛两个阶段，赛方分别提供了训练数据和测试数据。这部分的数据只是在内容上有不同，但是形式和任务并没有任何的变化。所以可能把两部分数据整合起来一块分析。由于赛后并没有提供测试集的结果，所以我们只能通过从训练集中留出部分数据作为验证集来评估算法的性能。

对赛方提供的样本进行分析，主要是统计各个类别的比例。

```python
# -*- coding:utf-8 -*-

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import jieba
import time
from tqdm import tqdm
import sys
import os
```

```python
# 导入 Label.csv 和 LabelSecond.csv 进行类别统计
Label_path = '../raw_data/Label.csv'
LabelSecond_path = '../raw_data/LabelSecond.csv'

with open(Label_path, 'rb') as inp:
    df_Label = pd.read_csv(inp, sep='\t')
    
with open(LabelSecond_path, 'rb') as inp:
    df_LabelSecond = pd.read_csv(inp, sep='\t')

df_label = pd.concat([df_Label, df_LabelSecond])
```

```python
def df_analyze(df):
    """对 Label 数据框进行分析：视角个数，情感倾向分布。"""
    plt.clf()
    SentenceId = df.SentenceId
    View= df.View
    Opinion = df.Opinion
    SentenceId_count = SentenceId.value_counts()  # 评论句子数量
    # **每个句子的 view 数量分布
    view_max = SentenceId_count.values[0]    # view 最多的数量
    view_min = SentenceId_count.values[-1]   # view 最少的数量
    view_mean = np.mean(SentenceId_count.values)
    plt.hist(SentenceId_count.values, bins=view_max-view_min+1)
    plt.xlabel('view number/sentence')
    plt.ylabel('sentence number')
    plt.title('view distribution')
    plt.grid()
    plt.show()
    # **情感倾向分布
    Opinion_count = Opinion.value_counts()
    labels = Opinion_count.index
    sizes = Opinion_count.values
    explode = (0.05,0,0)
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.2f%%',
        shadow=False, startangle=90)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.title('opinion distribution')
    plt.show()
    # **每个句子的视角数量
    print 'All %d sentences. \n \
    view_max = %d, view_min = %d, view_mean = %g' % (len(SentenceId_count), view_max, view_min, view_mean)
    # **语料中不同视角的个数
    print 'There are %d view-opinion pairs; %d different views.' % (len(View), len(set(View.values)))
```

```python
print '= ' * 10, 'Analyze Label.csv', '= ' * 10
df_analyze(df_Label)
```

```python
print '= ' * 10, 'Analyze LabelSecond.csv', '= ' * 10
df_analyze(df_LabelSecond)
```

```python
print '= ' * 10, 'Analyze all the training date.', '= ' * 10
df_analyze(df_label)
```

## 分析结果： 
1.对于平均每个句子中包含的视角数量，两个数据都差不多。最后 每个句子大概含有 1.42 个视角。

2.对于视角的情感倾向，初赛和复赛给出的数据分布差距很大，复赛中有 85% 都是中性的视角。将两部分数据整合后中性视角仍然占了 69.41%，而负向视角才占了 7.44% ，所以数据还是比较不均衡的。

3.在初赛中有 1956 个不同的视角，复赛中有 1630 个不同的视角。将两个语料合起来分析发现不同的视角个数达到了 3030, 也就是初赛和复赛中仅有500 多个视角是一样的。所以这两部分数据要么分布差异很大，要么是标注的方式有出入。如果是后者的话，也没有太好的办法解决。但是既然都是汽车评论语料，如果分布差异大的话也不影响，正好验证算法的泛化能力。
